```python
import matplotlib.pyplot as plt
import numpy as np

data = np.array([
    [8.46666371, 5.09503086],
    [8.52442939, 5.55359076],
    [8.57989524, 5.57739563],
    [8.74932926, 5.81331286],
    [8.75484946, 5.80996909],
    [8.75881814, 5.87410311]
])

x = [2**i for i in range(3,3 + len(data[:,0]))]
y1 = data[:,0]/data[-1,0]
y2 = data[:,1]/data[-1,1]

plt.figure(figsize=(8, 5))
plt.plot(x, y1, marker='o', label="Mean")
plt.plot(x, y2, marker='o', label="Std")
plt.fill_between(x, 0.98*np.ones(len(y1)), 1.02*np.ones(len(y1)), alpha = 0.5, color='grey', label="Â±2% range")
plt.xscale('log', base=2)
plt.legend()

plt.xlabel('Number of samples')
plt.ylabel('Normalised indicators')
plt.grid(True)

plt.show()
```
